# Handling missing values

```python
# Handling missing values for age column
imputer = SimpleImputer(strategy='median')

imputer.fit(df[["Age"]])
df["Age"] = imputer.transform(df[["Age"]])
```

```python
# Handling missing values for Fare column
imputer = SimpleImputer(strategy='mean')

imputer.fit(df[["Fare"]])
df["Fare"] = imputer.transform(df[["Fare"]])
```

```python
# Handling missing values for Embarked column
df["Embarked"].fillna("S", inplace = True)
```

```python
df
```

# Converting categorical columns into numerical

```python
le = LabelEncoder()

df["Sex"] = le.fit_transform(df["Sex"])
```

```python
df= pd.get_dummies(df, columns = ["Embarked"], prefix = ["Embarked"])

le = LabelEncoder()

df["Embarked_C"] = le.fit_transform(df["Embarked_C"])
df["Embarked_Q"] = le.fit_transform(df["Embarked_Q"])
df["Embarked_S"] = le.fit_transform(df["Embarked_S"])
```
